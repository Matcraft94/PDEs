# Oscilador Armonico
La formulacion de esta ecuacion es la siguiente:
$$\ddot{x}+\dfrac{c}{m}\ddot{x}+\dfrac{k}{m}x=0$$
donde $m$ es la masa, $k$ es la constante del resorte y $c$ es el coeficiente de friccion.

```python
# Paquetes necesarios
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definicios una funcion que soluciona el problema dados parametros
def solve_harmonic(m, k, c = 0, x0=5, v0=0, span=[0,10]):
    
    def harmonic(t, q):
        return [q[1], -k*q[0]/m - c*q[1]/m]
    
    return solve_ivp(harmonic, span, [x0, v0], dense_output=True)
```

```python
# Configuracion de parametros
m = 1
k = 1
c = 0
x0 = 2.
v0 = 0.
span = [0,15]

# Solucion del problema
solution = solve_harmonic(m,k,c,x0,v0,span)
```

```python
# Mostramos la solucion
t = np.linspace(0, 15, 300)
z = solution.sol(t)

plt.plot(t, z.T)
plt.xlabel('t')
plt.legend(['x', 'v'], shadow=True)
```

```python
plt.plot(z[0], z[1])
plt.arrow(2, 0, 0, 0.2, shape='full', lw=0, length_includes_head=True, head_width=0.2)
plt.title('Oscilador Armonico - Espacio de Fase')
```
